## Importar librerias y cargar data

```python
# Importarmos pandas, numpy y missingno
import pandas as pd
import numpy as np
import missingno as msno
```

```python
# almacenar los daots en un dataFrame
df2 = pd.read_csv('autos.csv',sep=';')
# mostrar las ultimas 10 filas
df2.head(18)
```

## Identificar datos perdidos

```python
# Mostrar Total datos perdidos por columna
print('-'*50,'Total datos perdidos por columna')
print(pd.DataFrame(
    data=df2.isna().sum(),
    columns=['Datos Perdidos']
))
# Mostrar informaciÃ³n del dataFrame
print('-'*50,'Info del dataFrame')
print(df2.info())
# Mostrar grafico de datos perdidos
print('-'*50,'Grafico - datos perdidos')
print(msno.matrix(df2))
```

## Tratar datos perdidos

```python
col_num = ['Millas_Por_Galon','Caballo_De_Fuerza']
# a la columna de la lista se le rellenara con la media de la columna
df2[col_num] = df2[col_num].fillna(df2[col_num].median())
# volvemos a mostrar los datos perdidos por columna
print(pd.DataFrame(data=df2.isna().sum(),columns=['datos perdidos']))
# vemos la informacion del dataFrame
print(df2.info())
```

## Agregar las nuevas variables

```python
'''
cada funcion recibe como parametro dos series, y retorna una nueva para ser asignado al dataFrame
la serie 1 es millas por galon y la serie 2 es nÃºmero de cilindros
'''
def c1(s1,s2):
  resultado = s1/s2 
  # como resultado tenemos una serie
  # a cada elemento de la serie aplicamos logaritmo con base 10
  resultado = resultado.apply(lambda x:np.log10(x))
  return resultado.round(2)
def c2(s1,s2):
  # raiz cuadrada de s1 * exp(s2) dividido en 200
  resultado = np.sqrt(s1)*np.exp(s2)/200
  return resultado.round(2)
def c3(s1,s2):
  # si s1 es mayor a s2 retornamos 5 caso contrario 3
  resultado = np.where(s1>s2,5,3)
  return resultado
def c4(s1,s2):
  # 1 / logaritmo natural de s1/s2 *100
  resultado = 1/np.log(s1/s2)*100
  return resultado.round(2)
def c5(s1,s2):
  # s2 elevado a 2 dividido entre s1
  resultado = s2**2/s1
  return resultado.round(2)
```

```python
# aÃ±adiendo variables con lambda
df = (
    df
    .assign(col_1 = lambda x: c1(x.Millas_Por_Galon,x.Cilindros)) # pasamos como parametro millas por galon y cilindros
    .assign(col_2 = lambda x: c2(x.Millas_Por_Galon,x.Cilindros))
    .assign(col_3 = lambda x: c3(x.Millas_Por_Galon,x.Cilindros))
    .assign(col_4 = lambda x: c4(x.Millas_Por_Galon,x.Cilindros))
    .assign(col_5 = lambda x: c5(x.Millas_Por_Galon,x.Cilindros))
)
df.head(5)
```

## FunciÃ³n categorizadora

```python
def var_cualitativa(serie1):
  '''
  crea variable cualitativa ordinaria a partir de variable numerica

  ---> Parametros
  serie1 = una serie con datos numericos en la que aplicaremos algunos criterios

  ---> Return
  una serie con datos cualitativos

  ---> Ejemplo
  serie1 = 1,2,3,4
  var_cualitativa(serie1)
  resultado = Bajo, Medio, Medio, Alto
  
  '''
  calculo1 = serie1.median()+1*serie1.std()
  calculo2 = serie1.median()-1*serie1.std()
  serieResultado = serie1.apply(lambda x: 'Alto' if x>calculo1 else('Medio' if (x > calculo2 and x < calculo1) else('Bajo' if x < calculo2 else() )))
  return serieResultado
```

## AÃ±adir nueva columna

```python
df3 = (
    df3
    # asignar al dataFrame la columna Aceleracion_Categoria con datos cualitativos
    .assign(Aceleracion_Categoria = lambda x: var_cualitativa(x.Aceleracion))
)
# mostrar las ultimas 7 filas
df3.tail(7)
```
